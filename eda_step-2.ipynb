```python
# df['target_runs'].dtype
 
def cal_team_score(data):
    #team scores
    first_batting = data['target_runs'] - 1
    second_batting = data['target_runs'] - data['result_margin']

    # #identify the teams bat and ball first and won the toss
    # if data['toss_decision'] == 'bat':
    #     first_batting_team = data['toss_winner']
    # else:
    #     first_batting_team = data['team1'] if data['toss_winner'] == data['team2'] else data['team2'] 

    #------------------------------------------------------------------------------------------------------------
    if data['team1'] == data['toss_winner']:
        if data['toss_decision'] == 'bat':
            team1_score = first_batting
            team2_score = second_batting
        else:
            team1_score = second_batting
            team2_score = first_batting
    else:  # team2 won toss
        if data['toss_decision'] == 'bat':
            team2_score = first_batting
            team1_score = second_batting
        else:
            team2_score = second_batting
            team1_score = first_batting

    #------------------------------------------------------------------------------------------------------------
    # if first_batting_team == data['team1']:
    #     team1_score = first_batting
    #     team2_score = second_batting
    # else:
    #     team1_score = second_batting
    #     team2_score = first_batting

    # Assign score to team1

    return pd.Series({'t2_score': team2_score, 't1_score': team1_score})
df[['t2_score','t1_score']] = df.apply(cal_team_score, axis = 1)

# df['1st_innings_runs'] = np.where(
#     df['toss_decision'] == 'bat',
#     bat1,
#     df['target_runs'] - df['result_margin'] 
# )
    
```
